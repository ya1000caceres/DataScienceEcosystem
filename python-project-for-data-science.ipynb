```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots


```

```python
# Crear un objeto ticker para Tesla
tesla = yf.Ticker("TSLA")

# Obtener el historial de datos del mercado para Tesla para el período máximo
tesla_data = tesla.history(period="max")

# Reiniciar el índice del DataFrame
tesla_data.reset_index(inplace=True)

# Mostrar las primeras cinco filas del DataFrame
print("Datos de Acciones de Tesla (TSLA):")
print(tesla_data.head())
```

```python
# URL para los datos de ingresos de Tesla
url_tesla = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Descargar el contenido de la página web
html_data_tesla = requests.get(url_tesla).text

# Parsear el contenido HTML con BeautifulSoup
soup_tesla = BeautifulSoup(html_data_tesla, 'html.parser')

# Encontrar la tabla con los datos de ingresos
# NOTA: La tabla de interés en esta página es la segunda.
tables = soup_tesla.find_all('table')
for index, table in enumerate(tables):
    if ("Tesla Quarterly Revenue" in str(table)):
        table_index = index

tesla_revenue = pd.read_html(str(tables[table_index]))[0]

# Renombrar las columnas y limpiar los datos
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]


# Mostrar las últimas cinco filas del DataFrame
print("Datos de Ingresos de Tesla (TSLA):")
print(tesla_revenue.tail())
```

```python
# Crear un objeto ticker para GameStop
gme = yf.Ticker("GME")

# Obtener el historial de datos del mercado para GameStop
gme_data = gme.history(period="max")

# Reiniciar el índice del DataFrame
gme_data.reset_index(inplace=True)

# Mostrar las primeras cinco filas del DataFrame
print("Datos de Acciones de GameStop (GME):")
print(gme_data.head())
```

```python
# URL para los datos de ingresos de GameStop
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Descargar el contenido de la página web
html_data_gme = requests.get(url_gme).text

# Parsear el contenido HTML
soup_gme = BeautifulSoup(html_data_gme, 'html.parser')

# Encontrar la tabla con los datos de ingresos de GME
tables_gme = soup_gme.find_all('table')
for index, table in enumerate(tables_gme):
    if ("GameStop Quarterly Revenue" in str(table)):
        table_index_gme = index

gme_revenue = pd.read_html(str(tables_gme[table_index_gme]))[0]

# Renombrar las columnas y limpiar los datos
gme_revenue.columns = ['Date', 'Revenue']
gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace(',|\$',"", regex=True)
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]

# Mostrar las últimas cinco filas del DataFrame
print("\nDatos de Ingresos de GameStop (GME):")
print(gme_revenue.tail())
```

```python
def make_graph(stock_data, revenue_data, stock_name):
    """
    Crea un gráfico con dos ejes Y para mostrar el precio histórico de las acciones y los ingresos.
    """
    fig = make_subplots(rows=1, cols=1, shared_xaxes=True, specs=[[{"secondary_y": True}]])
    
    # Asegurarse de que las columnas de fecha estén en formato datetime
    stock_data["Date"] = pd.to_datetime(stock_data["Date"], infer_datetime_format=True)
    revenue_data["Date"] = pd.to_datetime(revenue_data["Date"], infer_datetime_format=True)
    
    # Añadir la serie de datos de acciones
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], name="Precio de Cierre"), secondary_y=False)
    
    # Añadir la serie de datos de ingresos
    fig.add_trace(go.Scatter(x=revenue_data['Date'], y=revenue_data['Revenue'].astype("float"), name="Ingresos"), secondary_y=True)
    
    # Configurar títulos y etiquetas
    fig.update_layout(
        title_text=f"Precio Histórico de Acciones e Ingresos de {stock_name}",
        xaxis_title="Fecha"
    )
    fig.update_yaxes(title_text="Precio de Cierre (USD)", secondary_y=False)
    fig.update_yaxes(title_text="Ingresos (USD Millones)", secondary_y=True)
    
    fig.show()

# Llamar a la función para graficar los datos de Tesla
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
# Llamar a la función para graficar los datos de GameStop
make_graph(gme_data, gme_revenue, 'GameStop')
```
